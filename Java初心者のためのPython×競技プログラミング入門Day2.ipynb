## section3 コレクション

 - list

````python
arr = []
arr.append(1)
arr.append(2)
arr.append(3)

print(str(arr[0]) + str(arr[1]) + str(arr[2]))

#便利なので載せておきます。簡単に言うと、joinはlistをまとめて文字列にしてくれるもの、
#mapはlistの各値に一定の処理をしてくれるもの(今回は全部strにしてる)です。
print(''.join(map(str, arr)))
````

````python
arr = []
arr.append(1)
arr.append('Hello, World')
arr.append([1, 2, 3, 4, 5])
print(arr)
````

````python
arr = [1, 2, 3, 4, 5]

#追加：リストの末尾に指定した値が追加できます。
arr.append(6)
print(arr)
````

````python
arr = [1, 2, 3, 4, 5]

#挿入：リストの好きな位置に指定した値が追加できます。
arr.insert(0, 100)#一つ目の引数にindex、二つ目の引数が入れたい値です。
print(arr)
````

````python
arr = [1, 2, 3, 4, 5]

#削除(取り出し)
arr.pop(0)
print(arr)

#削除時に値が返ってきます
print(arr.pop(0))
print(arr)
````

 - dict(辞書)

````python
arr = {}#listは[]、dict（辞書)は{}をつかいます。
arr['apple'] = 100
arr['orange'] = 300

print(arr.items())
print(arr["apple"])
print(arr.keys())
print(arr.values())
````

````python
arr = {'apple': 100, 'orange': 300}

#追加
arr['banana'] = 500
print(arr.items())

#削除
del arr['apple']
print(arr.items())
````

````python
arr = {'apple': 100, 'orange': 300}

#便利なので載せておきます。orangeが存在すればHello, Worldします。
if 'orange' in arr:
  print('Hello, World')
````

## section4 制御構文

#### if

````python
if 2 > 1:
    print('Hello, World')

n = 5
if n > 6:
    print('n > 6')
elif n < 4 and n != 5:
    print('n < 4 かつ n != 5')
else:
    print('n == 5')
````

#### for

````python
#引数に一つだけ値(n)を渡したとき、0以上n未満の整数が生成されます。
for i in range(5):
    print(i)
````

````python
#引数にnとmを渡せば、n以上m未満の整数が生成されます。
for i in range(3, 8):
    print(i)
````

````python
#引数にnとmとlを渡せばnからmまでの整数がlの幅で生成されます。
for i in range(0, 5, 2):
    print(i)
````

````python
#逆順にもできます。
for i in range(10, 0, -1):
    print(i)
````

````python
#iにlistの値を入れたい場合
arr = [100, 200, 300]
for i in arr:
    print(arr)
````

````python
#listの長さ分だけ実行したい場合
arr = [100, 200, 300]
for i in len(arr):
    print('Hello, World')
````

#### while

````python
i = 0
while(i < 5):
  i += 1
  print('Hello, World')

while(True):
  i += 1
  print('i += 1')
  if i > 10:
    break
````
